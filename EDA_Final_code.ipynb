```python
# checking duplicates

duplicate = df.duplicated()
df = df.drop_duplicates(subset=['Museum ID'])
df[df.duplicated(['Museum ID'])]
print(duplicate.sum())
```

```python
#Drop unimportant columns
df.drop(['Legal Name', 'Alternate Name', 
         'Street Address (Physical Location)', 'City (Physical Location)', 
         'State (Physical Location)', 'Employer ID Number'], inplace=True, axis=1)
df
```

```python
#Rename columns to facilitate analysis
df = df.rename(columns={'City (Administrative Location)': 'City',
                        'Street Address (Administrative Location)':'Street Address',
                        'State (Administrative Location)':'State',
                        'Zip Code (Administrative Location)':'Zip code(AL)',
                        'Zip Code (Physical Location)':'Zip code(PL)'})
df.columns
```

```python
df.columns = df.columns.str.replace(' ', '')
```

```python
#To make sure the columns are changed
df.columns
```

```python
df.drop(['InstitutionName', 'Zipcode(PL)'], inplace=True, axis=1)
df
```

```python
# fill missing values with the sample mean 
df['Revenue'].fillna((df['Revenue'].mean()), inplace=True)
```

# 4.Visualization

```python
def museums(x):
    if x=='ARBORETUM, BOTANICAL GARDEN, OR NATURE CENTER':
        x='NATURE CENTER'
    elif x=='SCIENCE & TECHNOLOGY MUSEUM OR PLANETARIUM':
        x='SCIENCE & TECHNOLOGY MUSEUM'
    elif x=='ZOO, AQUARIUM, OR WILDLIFE CONSERVATION':
        x='WILDLIFE CONSERVATION'    
    return x
df['MuseumType']=df['MuseumType'].map(lambda x: museums(x))
```

```python
type_rev = df.groupby(['MuseumType']).agg({'Revenue':'sum'})
type_rev = type_rev.sort_values(by='Revenue', ascending=False)
type_rev.plot(kind='bar', figsize=(11,6), label='cold');
plt.savefig("MuseumsType1.jpg")
```

```python
sCity = pd.DataFrame(df['State'].value_counts())
sCity.columns = ['Count']
sCity['State'] = sCity.index.tolist()
sCity.sort_values(by="Count",ascending=False)
sCity = sCity.reset_index(drop=True)
sCity
```

```python
def museums(x):
    if x=='ARBORETUM, BOTANICAL GARDEN, OR NATURE CENTER':
        x='NATURE CENTER'
    elif x=='SCIENCE & TECHNOLOGY MUSEUM OR PLANETARIUM':
        x='SCIENCE & TECHNOLOGY MUSEUM'
    elif x=='ZOO, AQUARIUM, OR WILDLIFE CONSERVATION':
        x='WILDLIFE CONSERVATION'    
    return x
df['MuseumType']=df['MuseumType'].map(lambda x: museums(x))
```

```python
# Let's count the number of museum types available

plt.figure(figsize=(10,5))
sns.countplot(data=df, x='MuseumType', palette="Set3")
plt.title('Counts of Muesum Type', fontsize=15,color='gray')
plt.xlabel('MuseumType',color='gray')
plt.ylabel("Count",color='gray')
plt.xticks(rotation=90);
plt.savefig("Counts of Muesum Type.jpg")
```

We notice Natural history Museum that are very low % compared to the rest of the museum types

```python
#Highest state has number of museums in the data frame.
sns.set(style='whitegrid')
plt.figure(figsize=(10,15))
sns.countplot(y = 'State',data = df,order = df['State'].value_counts().index)
plt.ylabel('State',fontsize = 15, weight = 'bold',color='gray');
plt.xlabel('Museum number',fontsize = 15, weight = 'bold',color='gray');
plt.title('Highest state has number of museums',fontsize = 20,color='gray')
plt.show() #show the plot
plt.savefig("Highest state has number of museums.jpg")
```

Based on the above countplot, it seems that the number of art museums varies much more with a state's population than the number of science museums.)
